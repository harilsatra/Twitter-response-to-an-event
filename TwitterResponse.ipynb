### Grouping the locations according to the frequency

**Step1: Merge the two columns namely latitude and longitude into one single column.**

**Step2: Store this list in a table. The table will calculate the frequency of each element of the list and return only one instance of that element along with its frequency.**

**Step3: Split the merged latitude and longitude returned by the table using the function strsplit.**

**Step4: Transform the data returned from the table into 3 columns two of them being the separated latitude and longitude and the third being th corresponding frequency.** 

```r
lonlat <- paste(geocode_final.df$lon, geocode_final.df$lat)
table <- table(lonlat)
x <- as.data.frame(table(lonlat))
#Reference: http://stats.stackexchange.com/questions/6806/splitting-a-numeric-column-for-a-dataframe
df_split<- strsplit(as.character(x$lonlat), split=" ")
x <- transform(x, lon= sapply(df_split, "[[", 1),lat= sapply(df_split, "[[", 2))
head(x)
```

##### Write the above generated data frame into another CSV file 

```r
write.csv(x, file="lonlat_withFreq.csv")
```

##### Read all the geocodes from the CSV file into a dataframe.

```r
y <- read.csv("lonlat_withFreq.csv") 
```

#### Creating the map and plotting the loactions

#### Reference: https://blog.dominodatalab.com/geographic-visualization-with-rs-ggmaps/

```r
#Get the geo code of United States and save it in a variable
usa_center = as.numeric(geocode("United States"))
```

```r
#Using the geocode to get the map of USA
USAMap = ggmap(get_googlemap(center=usa_center, scale=2, zoom=4), extent="panel")
```

```r
#Displaying the map and potting the locations on the map with the size of the blob a multiple of the frequency of that location in the data.
USAMap + geom_point(aes(x=lon, y=lat), data=y, col="purple", alpha=0.55, size=y$Freq*0.4)
```
